### Write-up:

After completing the project, I have created a logistic regression model that predicts which users will become Adopters with 95.6% accuracy. In all of the created models (logistic regression, decision tree, random forest), the main predictor of Adoption was the difference in time between the account creation and the latest login. This is not surprising, given a full 79% of non-Adopters never log in again after creating their account. If a look-back is not allowed, the best model is the decision tree classifier, which has 82.5% accuracy. The major predictors for future user adoption are not being invited to join someone's "Personal Projects" but instead having limited access as a "guest." 

The results were surprising in many ways. Specifically, inviting more people decreases users' chances of becoming Adopters. Also, giving a fake email address has no bearing on user adoption, nor does opting in to the mailing list nor getting regular emails.

The first thing we need to do is import the data and make sure all the datetimes are correct.

```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt

# suppress warnings for for neater appearance
import warnings
warnings.filterwarnings('ignore')

# import data
engage = pd.read_csv('takehome_user_engagement.csv', infer_datetime_format=True, parse_dates=['time_stamp'])
users = pd.read_csv('takehome_users.csv', infer_datetime_format=True, parse_dates=['creation_time'], encoding='latin1')
users['last_session_creation_time'] = pd.to_datetime(users['last_session_creation_time'], unit='s')
```

Now that the dat is all in and looks good, we need to identify which users are "adopted users" ("Adopters"). Users are Adopters if they log in at least three different days in a 7-day period. I would love to do this with some kind of multi-index timestamp difference count, but I can't figure out how to accomplish that. Instead, I'll loop through. Not ideal, but not too slow. We'll need to resample by user and day to discard multiple logins for the same user on the same day. Then for each user, we'll get the difference in days between each of the logins. For that list of day-differences, we'll just add up every ordered pair and if the sum is less than 7, we can add that user to the set of adopters.

```python
adopters = set()
for i in engage.user_id.unique():
    ts = engage.loc[engage.user_id==i, 'time_stamp']
    d = list()
    for j in range(len(ts)):
        try:
            d.append((ts.iloc[j+1] - ts.iloc[j]).days)
        except:
            pass
    if len(d) > 1:
        for k in range(len(d)):
            try:
                if sum(d[k:k+2]) < 7:
                    adopters.update([i])
                    continue
            except:
                pass
```

Let's add a boolean column in the users dataset for Adopters, then make some more columns that can help us understand the data:
1. diff_time: number of days between creation_time and last_session_creation_time
2. domain: email domain

```python
users = users.set_index('object_id')
users['adopter'] = False
users.loc[adopters, 'adopter'] = True
users['diff_time'] = (users['last_session_creation_time'] - users['creation_time'])
users['diff_time'] = users.set_index('diff_time').index.days
users['domain'] = users.email.apply(lambda x: re.split('@', x)[1])
users.head()
```

It turns out the invited_by_user_id column has a lot of NaNs, so we should make an 'invited' boolean column. Also, a few thousand users don't have a last_session_creation_time value, meaning they signed up and never logged back in. We should make a 'no_login' boolean column based on this. We can also see that there are 1184 user domains and many of them look fake, so we need to make a 'fake email' column and see what else we can do with that.

```python
# add cols
users['no_login'] = pd.isnull(users['diff_time'])
users['is_invited'] = ~pd.isnull(users['invited_by_user_id'])

# look at domain counts
users.set_index('domain')[users.groupby('domain').email.count() > 1].groupby('domain').count()
```

```python
users.set_index('domain')[users.groupby('domain').email.count() > 5].groupby('domain').count()
```

```python
fake_domain = users.set_index('domain')[users.groupby('domain').email.count() < 5].groupby('domain').count().index.tolist()
users['dup_domain'] = users['domain']
users = users.reset_index().set_index('dup_domain')
users.loc[fake_domain, 'domain'] = 'fake'
users = users.reset_index()
users.drop('dup_domain', axis=1, inplace=True)
```

For these columns, because their numbers aren't meaningful in a regression sense, we'll break them into groups. Because we don't want to over-parameterize the model, we'll only add the non-first groups:
1. org_id: 0-20, 21-50, 51-100, 101+
2. invited_by_user_id: 0, 1-3, 4+

```python
# columns for users in groups by organization size
users = users.set_index('org_id')
users['org_size'] = '0-20'
users.loc[org_id_counts[(org_id_counts > 20) & (org_id_counts <=50)].index, 'org_size'] = '21-50'
users.loc[org_id_counts[(org_id_counts > 50) & (org_id_counts <=100)].index, 'org_size'] = '51-100'
users.loc[org_id_counts[org_id_counts > 100].index, 'org_size'] = '100+'
```

```python
# columns for users in groups by the size of the group they were invited with
users = users.reset_index().set_index('invited_by_user_id')
users['invited_with'] = '0'
users.loc[invited_counts[(invited_counts >= 0) & (invited_counts < 4)].index, 'invited_with'] = '1-3'
users.loc[invited_counts[invited_counts > 4].index, 'invited_with'] = '4+'
```

```python
users = users.reset_index().set_index('object_id')
users['invited'] = '0'
users.loc[invited_counts[(invited_counts >= 0) & (invited_counts < 4)].index, 'invited'] = '1-3'
users.loc[invited_counts[invited_counts > 4].index, 'invited'] = '4+'
```

Alright, it's modeling time. We'll do an 80/20 train/test split with 10-fold cross-validation. 

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LogisticRegression, ElasticNetCV, SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

# set up data for sklearn
users_d = pd.get_dummies(users, drop_first=True)
preds = ['opted_in_to_mailing_list',
 'enabled_for_marketing_drip',
 'no_login',
 'is_invited',
 'creation_source_ORG_INVITE',
 'creation_source_PERSONAL_PROJECTS',
 'creation_source_SIGNUP',
 'creation_source_SIGNUP_GOOGLE_AUTH',
 'domain_fake',
 'domain_gmail.com',
 'domain_gustr.com',
 'domain_hotmail.com',
 'domain_jourrapide.com',
 'domain_yahoo.com',
 'org_size_100+',
 'org_size_21-50',
 'org_size_51-100',
 'invited_with_1-3',
 'invited_with_4+',
 'invited_1-3',
 'invited_4+',
 'diff_time']
X_train, X_test, y_train, y_test = train_test_split(users_d.loc[:, preds], users_d.loc[:, 'adopter'], test_size=.2, random_state=1)
```

```python
# set parameters, grid search, logistic model, and fit
paramgrid = {'l1_ratio': [.01, .1, .3, .5, .7, .9, .99], 'alpha': [0.0001, 0.001, 0.01, 0.1]}

log_cv = GridSearchCV(SGDClassifier(loss='log', 
                                    penalty='elasticnet',
                                    #max_iter=1000,
                                    random_state=1),
                      param_grid=paramgrid,
                      cv=10)

log_cv.fit(X_train, y_train)

# set up decision tree and random forest models
dt = DecisionTreeClassifier(random_state=1)
dt_cv = cross_val_score(dt, X_train, y_train, cv=10)
dt.fit(X_train, y_train)

rf = RandomForestClassifier(random_state=1)
rf_cv = cross_val_score(rf, X_train, y_train, cv=10)
rf.fit(X_train, y_train)
```

```python
# Test models

# Logistic Regression
print("###Logistic Regression###")
print("Test Accuracy: ", round(log_cv.score(X_test, y_test), 6), '\n')
print("Test Confusion Matrix:")
cm = pd.DataFrame(metrics.confusion_matrix(y_test, log_cv.predict(X_test)), columns=['Predicted: Non-Adopter', 'Predicted: Adopter'])
cm['True'] = ['Non-Adopter', 'Adopter']
cm.set_index('True', inplace=True)
print(cm, '\n\n')

# Decision tree
print("###Decision Tree###")
print("Test Accuracy: ", round(dt.score(X_test, y_test), 6), '\n')
print("Test Confusion Matrix:")
cm = pd.DataFrame(metrics.confusion_matrix(y_test, dt.predict(X_test)), columns=['Predicted: Non-Adopter', 'Predicted: Adopter'])
cm['True'] = ['Non-Adopter', 'Adopter']
cm.set_index('True', inplace=True)
print(cm, '\n\n')

# Random Forest
print("###Random Forest###")
print("Test Accuracy: ", round(rf.score(X_test, y_test), 6), '\n')
print("Test Confusion Matrix:")
cm = pd.DataFrame(metrics.confusion_matrix(y_test, rf.predict(X_test)), columns=['Predicted: Non-Adopter', 'Predicted: Adopter'])
cm['True'] = ['Non-Adopter', 'Adopter']
cm.set_index('True', inplace=True)
print(cm)
```

The regssion model has the best accuracy, and because it's the easiest to interpret, we'll look at the coefficients.

```python
# Model coefficients
feature = X_train.columns.tolist()
feature.append('intercept')
coefficient = log_cv.best_estimator_.coef_[0].tolist()
coefficient.append(log_cv.best_estimator_.intercept_[0])
betas = pd.DataFrame({'feature':feature, 'coefficient':coefficient})
betas['exp'] = np.exp(betas['coefficient']) * (betas['coefficient'] / abs(betas['coefficient']))
betas.loc[~pd.isnull(betas.exp), :]
```
